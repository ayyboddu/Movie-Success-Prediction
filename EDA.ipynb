# Pre-Processing

```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file & save as dataframe
movies = pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
movies.rating.unique()
```

```python
#remove rows with missing values

movies = movies.dropna(axis=0)
```

```python
#remove movies not created in US, UK, or Canada

keep_countries = ['United States','United Kingdom','Canada']
movies = movies[movies['country'].str.contains('|'.join(keep_countries))]
```

```python
movies.country.unique()
```

```python
# check for repeated values

print(len(movies))
movies.nunique()
```

```python
#remove movies with the same name that were also released in the same year
movies = movies.drop_duplicates(
    subset = ['name','year'],
    keep = 'last').reset_index(drop =True)
```

```python
#there were no duplicates. only movies with the same name
print(len(movies))
movies.nunique()
```

```python
#check datatypes

movies.dtypes
```

```python
#export csv
movies.to_csv('movies_cleaned')
```
